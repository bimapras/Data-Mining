```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("spek_hp.csv")
data.head()
```

```python
data_baru = data.loc[:, ['clock_speed','int_memory','n_cores','ram','rating','kelayakan']]
data_baru
```

```python
data_baru.info()
```

```python
kelayakan_hp = dict(zip(data_baru.rating.unique(), data_baru.kelayakan.unique())) #mengembalikan nilai sesuai dengan argumen yang diberikan
kelayakan_hp
```

# Proses visualiasasi menggunakan 2 diagram berupa diagram lingkaran dan diagram grafik

```python
s = data_baru['kelayakan'].value_counts()

plt.figure(figsize=(10,10))
plt.pie(s, labels = s.index, autopct=lambda p:f'{p:.2f}%, {p*sum(s)/100 :.0f} units')
plt.title('Sebaran banyaknya handphone yang dapat menggunakan aplikasi berat')
plt.show()
```

```python
names = ['kelayakan']

for i in names :
    plt.figure(figsize=(7,4), dpi=90)
    sns.countplot(x=i, data=data_baru, hue= 'rating')
    plt.title('Grafik sebaran handphone yang dapat menggunakan aplikasi berat')
    plt.show()
```
