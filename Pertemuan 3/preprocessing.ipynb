Memisahkan data training dan test sebesar 0.2 atau 20%

```python
x = dataset_baru[['clock_speed','int_memory','n_cores','ram','rating']]
y = dataset_baru[['kelayakan']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=1)
```
